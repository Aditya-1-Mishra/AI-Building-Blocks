### Importing required libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([corpus])
```

```python
len(tokenizer.word_index) # this line of code will tell us how many words are there
```

```python
total_words = len(tokenizer.word_index)+1
```

### Tokenization of words

```python
input_sequence =[]
#for seq in corpus.split('.'):
#    if seq :
#      print(seq + '.')  
    
for seq in corpus.split('.'): 
    tokenized_sentence = tokenizer.texts_to_sequences([seq])[0]
    for word in range(1,len(tokenized_sentence)):
        input_sequence.append(tokenized_sentence[:word+1])
        
```

```python
print(len(input_sequence))
input_sequence
```

```python
max_val=0
for sentence in input_sequence:
    max_val = max(max_val,len(sentence))

print(f"max_val of a sequence is : {max_val}")
```

### Padding sequences

```python
padded_input_sequences = pad_sequences(input_sequence,maxlen=max_val,padding='pre')
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

### OHE -> one hot encoding

```python
y = to_categorical(y,num_classes=total_words) # since we have 467 unique words
```

```python
y.shape
```

### Model Building & Compilation

```python
model = Sequential()
model.add(Embedding(total_words,150,input_length=max_val-1))
model.add(LSTM(200,return_sequences=True))
model.add(LSTM(200))
model.add(Dense(total_words,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X,y,epochs=150)
```

### Plot

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
```


### Model evaluation on a sentence

```python
import time
import numpy as np
sentence = 'accelerationism is a'

for t in range(10):
    # tokenize the sentence
    tokenized_inp = tokenizer.texts_to_sequences([sentence])[0]

    # padding
    padded_inp = pad_sequences([tokenized_inp],maxlen=max_val,padding='pre')

    # predict
    pos = np.argmax(model.predict(padded_inp))

    for word,index in tokenizer.word_index.items():
        if index == pos:
          sentence = sentence + " " + word
          print(sentence)
          time.sleep(2)
```
